```python
from transformers import BitsAndBytesConfig, T5Tokenizer, T5ForConditionalGeneration
import torch
```

```python
# Setting up the BitsAndBytes config
# Using nf4 as the quant type
bnb_config = BitsAndBytesConfig(load_in_4bit=True, bnb_4bit_use_double_quant=True, bnb_4bit_quant_type="nf4", bnb_4bit_compute_dtype=torch.bfloat16)
```

```python
# Using flan-t5-xxl without quantization will weigh around 38gb vs 13gb
tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-xxl")
model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-xxl", device_map="auto", quantization_config = bnb_config)

input_text = "translate German to English: Wie alt sind Sie?"
input_ids = tokenizer(input_text, return_tensors="pt").input_ids.to("cuda")

outputs = model.generate(input_ids)
print(tokenizer.decode(outputs[0]))
```
